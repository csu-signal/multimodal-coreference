```python
import sys
sys.path.append('/home/huma97/anaconda3/envs/hp/lib/python3.9/site-packages')
from PIL import Image
import torch
import pandas as pd
```

```python
def names_sentences(file_name):
    df = pd.read_csv(file_name)
    
    # Access the mention_ids and sentences columns of the dataframe
    mention_ids = df.iloc[:, 1].tolist()
    sentences = df.iloc[:, 2].tolist()
    image_names = []
    for m_id in mention_ids:
        image_names.append(m_id.replace('.xml_','_'))
        
    return image_names,sentences
```

```python
dev_im_names, dev_sentences = names_sentences('dev_sent.csv')
train_im_names, train_sentences = names_sentences('train_sent.csv')
test_im_names, test_sentences = names_sentences('test_sent.csv')
```

```python
from diffusers import StableDiffusionPipeline

pipe = StableDiffusionPipeline.from_pretrained("CompVis/stable-diffusion-v1-4")
pipe.to("cuda")

```

```python
def generate_images(im_names,sentences,folder_name):
    for i,(names,sent) in enumerate(zip(im_names, sentences)):
        print(i)
        generator = torch.Generator("cuda").manual_seed(1024)
        image = pipe(sent, guidance_scale=7.5, num_inference_steps=15, generator=generator).images[0]
        image.save("Generated_Images/"+folder_name+"/"+names+".jpg")
```

```python
generate_images(train_im_names, train_sentences,"train")
```

```python
generate_images(test_im_names, test_sentences,"test")
```
